```python
for column_name in ["default","housing","loan","y"]:
    df[column_name] = df[column_name].replace({"yes":1,"no":0})
```

```python
df.head()
```

```python
sns.heatmap(round(df.corr(),2),cbar=True,annot=True)
```

```python
duration_sorted = df.sort_values(['duration'],ascending=False)
duration_sorted.head(20)
```

```python
duration_sorted[duration_sorted['housing']==0]
```

```python
df['marital'] = df['marital'].replace({"single":0,"devorced":1,"married":2})
```

```python
df = df[df['education'] != 'unknown']
df['education'] = df['education'].replace({"primary":1,"secondary":2,"tertiary":3})
```

```python
df.head(20)
```

```python
month_map = {'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6, 'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12}

# Replace month abbreviations with numerical values
df['month'] = df['month'].replace(month_map)
```

```python
df.head()
```

```python
sns.heatmap(round(df.corr(),2),cbar=True,annot=True)
```

```python
df = df.drop(columns='month_num',axis=1)
```

```python
df['contact'] = df['contact'].replace({"unknown":0,"telephone":0,"cellular":1})
df
```

```python
sns.heatmap(round(df.corr(),2),cbar=True,annot=True)
```

```python
df['poutcome'] = df['poutcome'].replace({"unknown":0,"failure":0,"other":0,"success":1})
df
```

```python
sns.heatmap(round(df.corr(),2),cbar=True,annot=True)
```

```python
sns.heatmap(round(df[['housing','contact','duration','poutcome','y']].corr(),2),cbar=True,annot=True)
```
