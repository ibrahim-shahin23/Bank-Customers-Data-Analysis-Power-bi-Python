```python
df['y'] = df['y'].replace({"yes":1,"no":0})
df.head()
```

```python
sns.violinplot(x='duration',data=df)
```

```python
# Calculate cumulative sum of subscribed for each duration
df['cumulative_subscribed'] = df.groupby('duration')['y'].cumsum()

# Create the line chart
plt.plot(df['duration'], df['cumulative_subscribed'], marker='o', color='blue')
sns.displot(df['duration'],bins=35,kde=True,color='red')

# Set labels and title
plt.xlabel('Duration')
plt.ylabel('Cumulative Subscribed')
plt.title('Cumulative Subscribed vs. Duration')

# Show the plot
plt.show()
```
